```python
GIT Commands 
```

```python
1. status >> 
Syntax :: git status 
    show the status of current repo

2. init
Syntax :: git init 
    initialised the empty repository. 

3. add
Sytax:: git add file_name
    add the files to the staging area

    to add all files at once. 
    command : git add .


4. rm 
Syntax:: git rm --cached file_name
    remove the files from staging are and put back to untracked area. 

    to remove multiple files
    git rm --cached file1 file2

    to remove all the files 
    command git reset HEAD -- 

5. commit
Syntax :: git commit -m "commit message"
    it add files and folder to repository. 


6. log 
Syntax : git log 
    show all the commit info

7. branch 
Snntax: git branch >> it will show list of all branches with * as active branch. 
        TO CREATE A BRANCH : git branch my-branch-name

8. switch:
Syntax: git switch branch-name>> switch to speficied branch-name
        git switch -c branch-name >> to create new branch and switch to new-branch


9. checkout: 
Syntax: git checkout branch-name >> switch to specified branch-name
        git checkout -b branch-name >> to create new branch and switch to new-branch

```

```python
GIT Terminologies: 

repository: folder with all the project file along with .git named folder.
HEAD : Head is pointer which always point to the latest commit. 
BRANCH : Copy of a repo from the latest commit. 
```
